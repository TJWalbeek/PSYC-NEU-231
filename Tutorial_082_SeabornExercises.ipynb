# Seaborn Exercises

1) From the "tips" dataset
    1. add a percentage tip column to the dataset
    2. visualize the relationship between tip percentage, with day of week and time 
    3. visualize tip ~ day, time, and number of people (which function allows three variables independent variables at once?)


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips['PercentTip'] = tips['tip']/tips['total_bill'] * 100
tips.head()
```

```python
sns.pointplot(x='day',y='PercentTip',hue='time',data=tips)
```

```python
#visualize tip ~ day, time, and number of people 
sns.catplot(x='day',y='tip',hue='time',col='size',data=tips,kind='point', col_wrap=3)
plt.show()
```

2) Find an interesting trend in the "exercise" dataset

```python
exercise = sns.load_dataset('exercise')
print(exercise.head())
print(exercise.describe(include='all'))
print(exercise['diet'].unique())
print(exercise['time'].unique())
print(exercise['kind'].unique())
```

```python
sns.catplot(x='kind',y='pulse',hue='diet',col='time',data=exercise,kind='point', col_wrap=3)
plt.show()
```

```python
# More involved example from Seaborn Docs
g = sns.PairGrid(exercise, y_vars="pulse",
                 x_vars=["diet", "time", "kind"],
                 height=5, aspect=.5)

# Draw a seaborn pointplot onto each Axes
g.map(sns.pointplot, scale=1.3, errwidth=4, color="xkcd:plum")
g.set(ylim=(75, 125))
```
